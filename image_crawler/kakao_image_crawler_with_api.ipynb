```python
import os
import urllib.request
import json
import requests
import datetime
from pathlib import Path
```

```python
savePath = './images/kakao_api'
if not os.path.exists(savePath):
    os.makedirs(savePath, exist_ok=True)
```

```python
def crawl_image(keyword):
    image_path = os.path.join(savePath, keyword)
    if not os.path.exists(image_path):
        os.makedirs(image_path, exist_ok=True)
    encText = urllib.parse.quote(keyword)

    for page in range(1, 51):
        url = f"https://dapi.kakao.com/v2/search/image?query={keyword}&page={page}"

        headers = {
            'Authorization': 'KakaoAK 6dffe05b507a85b2ef6e7e96ea4db97f',

        }
        req = requests.get(url, headers=headers)

        if req.status_code == 200:
            html = req.text
            html = json.loads(html)['documents']
            images = [el['image_url'] for el in html]
            
            for url in images:
                try:
                    # 이미지링크 확인
                    print(url)

                    # 저장 파일명 및 경로
                    file_name = Path(url).name
                    FileName = os.path.join(image_path, file_name)

                    # 파일명 출력
                    print('full name : {}'.format(FileName))

                    # 이미지 다운로드 URL 요청
                    urllib.request.urlretrieve(url, FileName)

                except urllib.error.HTTPError as e:
                    print('oh no... : ', e.reason)
                    pass

                except urllib.error.URLError as e:
                    print('oh no... : ', e.reason)
                    pass
                except OSError as e:
                    print('oh no... : ', e)
                    now = datetime.datetime.now()
                    nowDatetime = now.strftime('%Y%m%d_%H%M%S')
                    FileName = os.path.join(image_path, nowDatetime)
                    urllib.request.urlretrieve(url, FileName)
                    pass
    
            # 다운로드 완료 시 출력
            print("--------download succeeded--------")

        else:
            print("Error Code:" + str(req.status_code))
```

```python
crawl_image('동물원')
```
